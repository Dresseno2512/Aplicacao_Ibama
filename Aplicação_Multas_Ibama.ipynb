
Plotando o gráfico da evolução do número total de infrações aplicadas anualmente. Especificamente, some o número de infrações ocorridas em cada ano e faça um gráfico onde o eixo horizontal contém os anos e o vertical o número total de infrações em cada ano.



```python
data1.groupby('ANO_INFRA').size().plot()

plt.ylabel(' Quantidade de Infrações')
plt.xlabel("Ano")
plt.show()
```
